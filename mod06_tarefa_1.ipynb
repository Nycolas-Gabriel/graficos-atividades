#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python

# Calculando a média de APGAR5 por CONSULTAS
df_grouped = df.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

# Definindo o estilo do gráfico
sns.set(style="whitegrid")

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(data=df_grouped, x='CONSULTAS', y='APGAR5', palette='Blues_d')

# Ajustando os rótulos e o título
plt.title('Média de APGAR5 por Número de Consultas')
plt.xlabel('Número de Consultas')
plt.ylabel('Média de APGAR5')

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Suponha que o DataFrame se chame 'df'
# Definindo o estilo do gráfico
sns.set(style="whitegrid")

# Criando a figura e os eixos para 4 gráficos (2x2)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Gráfico 1: Frequência de SEXO
sns.countplot(data=df, x='SEXO', ax=axes[0, 0], palette='Blues_d')
axes[0, 0].set_title('Frequência de SEXO')

# Gráfico 2: Frequência de GESTACAO
sns.countplot(data=df, x='GESTACAO', ax=axes[0, 1], palette='Greens_d')
axes[0, 1].set_title('Frequência de GESTACAO')

# Gráfico 3: Frequência de ESTCIVMAE
sns.countplot(data=df, x='ESTCIVMAE', ax=axes[1, 0], palette='Oranges_d')
axes[1, 0].set_title('Frequência de ESTCIVMAE')

# Gráfico 4: Frequência de CONSULTAS
sns.countplot(data=df, x='CONSULTAS', ax=axes[1, 1], palette='Purples_d')
axes[1, 1].set_title('Frequência de CONSULTAS')

# Ajustando a organização dos gráficos
plt.tight_layout()

# Exibindo o painel com os 4 gráficos
plt.show()
```
